**Create a plot based on the Variation of the price range by the vehicle type?**


```python
pricev= df.groupby(by='vehicleType')['price'].mean()

pricev.plot(kind='bar')


```

**Find out Total count of vehicles by type available on ebay for sale.As well as create a visualization for the client?**


```python
import seaborn as sns
import matplotlib.pyplot as plt

vehicle_type_counts = df['vehicleType'].value_counts()

# Print the total count of vehicles by type
print("Total count of vehicles by type:")
print(vehicle_type_counts)

# Create a bar plot for visualization
plt.figure(figsize=(12, 6))
sns.countplot(x='vehicleType', data=df, order=vehicle_type_counts.index)
plt.title('Total Count of Vehicles by Type on eBay')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

 **Is there any relationship between dollar_price and kilometer?**

```python
correlation_coefficient = df['price'].corr(df['kilometer'])
```

```python
correlation_coefficient
```

```python
correlation_matrix = df[['price', 'kilometer']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap: Dollar Price vs. Kilometer')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kilometer', y='price', data=df)
plt.title('Scatter Plot: Dollar Price vs. Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('price')
plt.show()
```
