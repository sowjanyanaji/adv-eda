```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 7))
sns.countplot(y='brand', hue='vehicleType', data=df)
plt.title("Average Price for Vehicles based on Vehicle Type and Brand")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a stacked bar chart
sns.countplot(x='vehicleType', hue='gearbox', data=df)
# Set the title
plt.title(" Average price for vehicles based on the vehicleType,gearbor")
```
